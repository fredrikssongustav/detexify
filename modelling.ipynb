```python
def createArray(sample):
    M = np.ones([map_width,map_height], dtype=int)

    for k in sample:
        for j in k:
            if j[1] > proxy_width - 1:
                i1 = map_width - 1
            else: 
                i1 = j[1]*map_width/proxy_width
                
            if j[0] > proxy_height - 1 :
                i2 = map_height - 1
            else:
                i2 = j[0]*map_height/proxy_height
                
            M[int(i1)][int(i2)] = 0
    return M
```

```python
# Load all data

cur.close()
cur = conn.cursor()
cur.execute("SELECT key, strokes FROM samples")

samples = cur.fetchall()
```

```python
from keras.utils import to_categorical 
df = pd.DataFrame(samples)
df[1] = df[1].apply(lambda x: createArray(x))

X = np.concatenate(df[1].to_numpy()).reshape(df.shape[0],32,32,1)

cats,uniques = df[0].factorize()
y = to_categorical(cats[0])
```

```python
from keras.utils import to_categorical
i = int(X.shape[0] * 0.9)
x_train = X[0:i]
y_train = y[0:i]
#y_train = df[0][0:i]

x_test = X[i:]
y_test =  y[i:]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D

img_rows = 32
img_cols = 32

def createModel():
    model = Sequential()
    model.add(Conv2D(32, kernel_size=(3, 3),
         activation='relu',
         input_shape=(img_rows, img_cols, 1)))

    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1098, activation='softmax'))

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=200, verbose=1)

scores = model.evaluate(x_test, y_test, verbose=0)
print(scores)
```
